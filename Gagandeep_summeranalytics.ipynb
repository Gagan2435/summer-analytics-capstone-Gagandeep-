```python
# ====== Summer Analytics Capstone: Dynamic Pricing (Clean Colab Version) ======

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Upload your CSV
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('dataset.csv')

# Step 3: Preprocessing
df['Timestamp'] = pd.to_datetime(df['LastUpdatedDate'] + ' ' + df['LastUpdatedTime'],
                                 format='%d-%m-%Y %H:%M:%S')
df = df.sort_values('Timestamp').reset_index(drop=True)

# Step 4: Calculate Pricing
df['Price'] = 10 + (df['Occupancy'] / df['Capacity'])

# Step 5: Plot Pricing vs. Time
plt.figure(figsize=(12,6))
plt.plot(df['Timestamp'], df['Price'], marker='o', linestyle='-', color='navy')
plt.title('Dynamic Pricing for Urban Parking Lots')
plt.xlabel('Time')
plt.ylabel('Price ()')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 6: Save Plot for Report
plt.savefig('dynamic_pricing_plot.png')
print(" Plot saved as 'dynamic_pricing_plot.png'. Download this image for your report.")
```
